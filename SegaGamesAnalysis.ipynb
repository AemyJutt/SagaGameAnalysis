# Segment Into Smaller Unique Values

```python
# in this feature there are many unique values that we need to handle this ?
df['meta_score'].unique()
```

```python
# we create a funtion 'def' to minimize our unique values into smaller values
def segment_meta_score(meta_score):
    if meta_score >= 90:
        return "Excellent"
    elif meta_score >= 80:
        return "Good"
    elif meta_score >= 70:
        return "Average"
    elif meta_score >= 60:
        return "Below Average"
    else:
        return "Poor"
```

```python
df['meta_score'] = df['meta_score'].apply(segment_meta_score)
```

```python
#now this looks better than before unique values
df['meta_score'].unique()
```

```python
df.head()
```

```python
# in this feature there are many unique values that we need to handle this ?
df['platform'].unique()
```

```python
# we create a funtion 'def' to minimize our unique values into smaller values
def categorize_platform(value):
    if 'XONE' in value or 'X360' in value or 'XBOX' in value:
        return 'Xbox'
    elif 'PC' in value:
        return 'PC'
    elif 'PS4' in value or 'PS3' in value or 'PS2' in value:
        return 'PlayStation'
    elif 'Switch' in value:
        return 'Nintendo Switch'
    elif '3DS' in value:
        return 'Nintendo 3DS'
    else:
        return 'others'
```

```python
df['platform'] = df['platform'].apply(categorize_platform)
```

```python
#now this looks better than before unique values
df['platform'].unique()
```

```python
df.head()
```

```python
# in this feature there are many unique values that we need to handle this ?
df['user_score'].unique()
```

```python
df['user_score'] = df['user_score'].fillna(df['user_score'].median())
```

```python
# we create a funtion 'def' to minimize our unique values into smaller values
def segment_score(score):
    if score <= 2:
        return "Very Low"
    elif score <= 4:
        return "Low"
    elif score <= 6:
        return "Medium"
    elif score <= 8:
        return "High"
    else:
        return "Very High"
```

```python
df['user_score'] = df['user_score'].apply(segment_score)
```

```python
#now this looks better than before unique values
df['user_score'].unique()
```

```python
df.head()
```

```python
df['esrb_rating'] = df['esrb_rating'].fillna(df['esrb_rating'].mode()[0])
```

```python
#in this feature we did not understand these kind of unique values ?
df['esrb_rating'].unique()
```

```python
#so we google it and convert that to better understanding for us
def categorize_esrb_rating(rating):
    if rating == 'E':
        return "Everyone"
    elif rating == 'RP':
        return "Rating Pending"
    elif rating == 'T':
        return "Teen"
    elif rating == 'M':
        return "Mature"
    elif rating == 'E10+':
        return "Everyone 10+"
    elif rating == 'K-A':
        return "Kids to Adults"
    else:
        return "Early Childhood"
```

```python
df['esrb_rating'] = df['esrb_rating'].apply(categorize_esrb_rating)
```

```python
# now these unique values is better and understandable before
df['esrb_rating'].unique()
```

```python
df.head()
```

# Exploratary Data Analysis

```python
cat_vars = ['meta_score','title','platform','date']

num_cols = len(cat_vars)

fig , axs = plt.subplots(nrows=4,ncols=1,figsize=(15,5*4))
axs = axs.flatten()

for i , var in enumerate (cat_vars):
    top_values = df[var].value_counts().nlargest(20).index
    filter_df = df[df[var].isin(top_values)]
    sns.countplot(y=var,data=filter_df,ax=axs[i])
    axs[i].set_title(var)

if num_cols < len(axs):
    for i in range(num_cols, len(axs)):
        fig.delaxes(axs[i])

fig.tight_layout()
plt.show()
```
